# Problema de Ingeniería (Ejemplo básico)

Un ingeniero agrícola necesita calcular **el área de un terreno rectangular** para estimar la cantidad de fertilizante que debe aplicar.

La fórmula es:

[
Área = largo \times ancho
]

---

# 1. Definición del algoritmo (Entradas – Proceso – Salidas)

### Entradas

Datos que el usuario debe proporcionar.

* `largo` del terreno (m)
* `ancho` del terreno (m)

### Proceso

Operación matemática para obtener el resultado.

* Calcular
  `area = largo * ancho`

### Salida

Resultado que el programa mostrará.

* Área del terreno en **metros cuadrados (m²)**

---

# 2. Diseño del algoritmo (explicación lógica)

1. Solicitar al usuario el **largo del terreno**
2. Solicitar el **ancho del terreno**
3. Multiplicar largo por ancho
4. Mostrar el área calculada

---

# 3. Pseudocódigo

```text
INICIO

leer largo
leer ancho

area ← largo * ancho

mostrar "El área del terreno es:", area

FIN
```

---

# 4. Implementación en Python

```python
# Programa para calcular el área de un terreno rectangular
# Autor: Estudiante de Ingeniería

# Entrada de datos
largo = float(input("Ingrese el largo del terreno (m): "))
ancho = float(input("Ingrese el ancho del terreno (m): "))

# Proceso
area = largo * ancho

# Salida
print("El área del terreno es:", area, "m²")
```

---

# 5. Ejemplo de ejecución

Entrada del usuario:

```

Ancho = 10
```

Resultado:

```

```

### Un primer acercamiento al código

En ingeniería, rara vez trabajamos con valores aislados; siempre buscamos relacionar variables para obtener un resultado técnico.

Antes de mostrarte el código, pensemos en la **lógica**: si quisiéramos calcular el volumen de agua necesario para un reservorio rectangular, ¿qué datos necesitaríamos y qué operación haríamos?

Aquí tienes un ejemplo de cómo se vería esa lógica traducida a Python, siguiendo buenas prácticas (variables descriptivas y modularización):

````python
# Ejemplo: Cálculo del volumen de una estructura de almacenamiento
def calcular_volumen_reservorio(largo_m, ancho_m, profundidad_m):
    """Calcula el volumen en metros cúbicos."""
    volumen_m3 = largo_m * ancho_m * profundidad_m
    return volumen_m3

# Datos de entrada (Imagina que vienen de un sensor o un formulario de Streamlit)
largo = 10.5
ancho = 5.0
profundidad = 2.0

# Llamada a la función
resultado = calcular_volumen_reservorio(largo, ancho, profundidad)

print(f"El volumen del reservorio es: {resultado} m³")
````
