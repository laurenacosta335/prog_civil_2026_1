```python
# ENTRADAS: amcho y alto en metros
ancho = input(" ingrese ancho de la ventana en metros ")
ancho = float(ancho)
alto = input(" ingrese el alto de la ventana en metros ")
alto = float(alto)
# PROCESO: Calcula lm en pies y av en metros cuadrados
lm = 2 * (ancho + alto) * 3.25
av = (ancho * alto)* 2
# SALIDA: sale longitud de la madera (lm) en pies y área de (av) vidrio en metros cuadrados
print("Longitus de la madera = ",lm,"pies de longitud ")
print("Área de la ventana = ",av,"Metros cuadrados ")
```
