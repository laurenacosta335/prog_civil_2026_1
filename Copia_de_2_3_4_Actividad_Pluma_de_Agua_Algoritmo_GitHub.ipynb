## **Historia, ciencia y programación: la “pluma de agua”**

> **Relato de partida**  
> **¿Por qué los costeños le decimos “pluma” a la llave del agua?**  
>  
> En el Caribe colombiano, abrir la “pluma” no es solo una acción cotidiana: es una expresión cargada de historia, técnica y cultura. Aunque algunos se sorprendan o incluso se burlen del término, lo cierto es que decir “pluma” en vez de “llave” no es un error, sino una herencia viva que conecta nuestro lenguaje con siglos de evolución hidráulica.  
>  
> Mucho antes de que existieran los grifos modernos, el agua se distribuía mediante sistemas regulados con precisión. En España, por ejemplo, se usaba la **“pluma de agua”** como unidad de medida para calcular el caudal que se entregaba a cada hogar. Aproximadamente, una pluma equivalía a **0.025 litros por segundo**. En algunos lugares, ese flujo se regulaba mediante orificios calibrados y dispositivos llamados **plumeros**, que permitían controlar la salida del agua.  
>  
> Durante la época colonial, muchas de estas prácticas técnicas cruzaron el Atlántico y llegaron a ciudades como Cartagena, Barranquilla y Santa Marta. Con el tiempo, el término “pluma” dejó de ser solamente técnico y pasó al habla cotidiana como sinónimo de la llave del agua.  
>  
> Así, una palabra del uso diario conserva una memoria histórica, tecnológica y cultural.

## 2. Situación problema

Un hogar recibe una cierta cantidad de **plumas de agua**.  
Se sabe que:

- **1 pluma = 0.025 litros por segundo**

Con esta información, el estudiante debe construir un programa que permita calcular:

1. el caudal total en litros por segundo;
2. el consumo diario en litros;
3. el consumo mensual en litros (suponiendo 30 días);
4. una clasificación del consumo:
   - **Bajo**: menos de 1000 litros por día
   - **Medio**: entre 1000 y 3000 litros por día
   - **Alto**: más de 3000 litros por día

- transformar el caudal a consumo diario;

### 3.3 Salida

Complete el siguiente bloque de código a partir del algoritmo anterior.

````python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON
# ==========================================================

# 1. Solicitar al usuario el número de plumas
plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema
CAUDAL_PLUMA = 0.025  # litros por segundo
SEGUNDOS_DIA = 24 * 60 * 60  # segundos en un día
DIAS_MES = 30 # suponiendo un mes de 30 días

# 3. Calcular el caudal total
caudal_total = plumas * CAUDAL_PLUMA

# 4. Calcular el consumo diario
consumo_diario = caudal_total * SEGUNDOS_DIA

# 5. Calcular el consumo mensual
consumo_mensual = consumo_diario * DIAS_MES

# 6. Clasificar el consumo
nivel_consumo = ""
if consumo_diario < 1000:
    nivel_consumo = "Bajo"
elif 1000 <= consumo_diario <= 3000:
    nivel_consumo = "Medio"
else:
    nivel_consumo = "Alto"

# 7. Mostrar resultados
print(f"\n--- Resultados del Consumo de Agua ---")
print(f"Número de plumas ingresadas: {plumas}")
print(f"Caudal total: {caudal_total:.4f} litros/segundo")
print(f"Consumo diario: {consumo_diario:.2f} litros/día")
print(f"Consumo mensual (aprox. 30 días): {consumo_mensual:.2f} litros/mes")
print(f"Clasificación del consumo: {nivel_consumo}")
print(f"-------------------------------------")
````

### 6. Versión Comentada del Código (sin cambiar su lógica)

```python
# ==========================================================
# ACTIVIDAD: MODELAR LA "PLUMA DE AGUA" CON PYTHON
# ==========================================================

# 1. Solicitar al usuario el número de plumas
# Se pide al usuario que ingrese la cantidad de "plumas" de agua.
# La entrada se convierte a tipo flotante (float) para permitir valores decimales.
plumas = float(input("Ingrese el número de plumas: "))

# 2. Definir constantes del problema
# CAUDAL_PLUMA: Caudal estándar de una sola pluma en litros por segundo.
CAUDAL_PLUMA = 0.025  # litros por segundo

# SEGUNDOS_DIA: Número total de segundos en un día (24 horas * 60 minutos/hora * 60 segundos/minuto).
SEGUNDOS_DIA = 24 * 60 * 60  # segundos en un día

# DIAS_MES: Número de días considerados en un mes para cálculos (se asume 30 días).
DIAS_MES = 30 # suponiendo un mes de 30 días

# 3. Calcular el caudal total
# Se calcula multiplicando el número de plumas por el caudal de una sola pluma.
# El resultado es el caudal total en litros por segundo.
caudal_total = plumas * CAUDAL_PLUMA

# 4. Calcular el consumo diario
# Se obtiene multiplicando el caudal total (L/s) por la cantidad de segundos en un día.
# El resultado es el consumo total en litros por día.
consumo_diario = caudal_total * SEGUNDOS_DIA

# 5. Calcular el consumo mensual
# Se calcula multiplicando el consumo diario por el número de días en un mes.
# El resultado es el consumo total en litros por mes.
consumo_mensual = consumo_diario * DIAS_MES

# 6. Clasificar el consumo
# Se inicializa una variable para almacenar la clasificación del consumo.
nivel_consumo = ""

# Se utiliza una estructura condicional (if-elif-else) para categorizar el consumo diario:
# Si el consumo diario es menor a 1000 litros, se clasifica como "Bajo".
if consumo_diario < 1000:
    nivel_consumo = "Bajo"
# Si el consumo diario está entre 1000 y 3000 litros (inclusive), se clasifica como "Medio".
elif 1000 <= consumo_diario <= 3000:
    nivel_consumo = "Medio"
# En cualquier otro caso (si el consumo diario es mayor a 3000 litros), se clasifica como "Alto".
else:
    nivel_consumo = "Alto"

# 7. Mostrar resultados
# Se imprimen todos los resultados de manera formateada para una mejor legibilidad.
# Los valores flotantes se formatean a dos o cuatro decimales según sea apropiado (e.g., ":.2f", ":.4f").
print(f"\n--- Resultados del Consumo de Agua ---")
print(f"Número de plumas ingresadas: {plumas}")
print(f"Caudal total: {caudal_total:.4f} litros/segundo")
print(f"Consumo diario: {consumo_diario:.2f} litros/día")
print(f"Consumo mensual (aprox. 30 días): {consumo_mensual:.2f} litros/mes")
print(f"Clasificación del consumo: {nivel_consumo}")
print(f"-------------------------------------")
```
