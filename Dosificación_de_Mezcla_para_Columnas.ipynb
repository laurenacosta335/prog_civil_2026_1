# ** Dosificación de Mezcla para Columnas**
**planteamiento**

Planteamiento:

Un ingeniero residente necesita calcular cuántos sacos de cemento se requieren para vaciar una columna de sección circular.

 Se sabe que para un concreto estándar de 210 kg/cm2 (3000 psi), se consumen 9.5 sacos de cemento por cada metro cúbico de mezcla.

**Requerimientos:**

1. Solicitar la altura de la columna en metros.
2. Solicitar el diámetro de la columna en metros.
3. Calcular el volumen de la columna:$$v = \pi * r^2 * h$$
. (Sugerencia: usar 3.1416
 o importar math.pi).
4. Calcular la cantidad de sacos:
$$sacos = v * 9.5$$

**Prueba del programa:**

Para una columna de 3 metros de alto y 0.4 metros de diámetro (r=0.2):

Volumen ≈0.377 m3.

Sacos necesarios ≈3.58

 (el estudiante verá aquí la importancia de los tipos decimales).


````python

# ENTRADA: Solicitar la altura de la columna en metros
# Solicitar el diámetro de la columna en metros.
h = float(input("Ingrese la altura de la columna en metros: "))
d = float(input("Ingrese el diámetro de la columna en metros: "))

# PROCESO: Calcula el radio r, calcula el volumen de la columna, calcula sacos
r = d/2
v = 3.1416 * r**2 * h
sacos = v * 9.5

# SALIDA: Imprimir el resultado
print("Volumen ≈", round(v, 3), "m3.")
print("Sacos necesarios ≈", round(sacos, 2), "sacos.")
````

```markdown
## Explicación del Código para el Cálculo de Sacos de Cemento

El siguiente código solicita la altura y el diámetro de una columna, calcula su volumen y determina la cantidad de sacos de cemento necesarios.

```python
h = float(input("Ingrese la altura de la columna en metros: "))
```
*   **`h = float(input("Ingrese la altura de la columna en metros: "))`**:
    *   `input("...")` muestra el mensaje entre comillas al usuario y espera que ingrese un valor.
    *   `float(...)` convierte la entrada del usuario (que es una cadena de texto) a un número decimal (punto flotante).
    *   El valor convertido se asigna a la variable `h` (altura).

```python
d = float(input("Ingrese el diámetro de la columna en metros: "))
```
*   **`d = float(input("Ingrese el diámetro de la columna en metros: "))`**:
    *   Similar a la línea anterior, solicita al usuario que ingrese el diámetro de la columna.
    *   Convierte la entrada a un número decimal.
    *   Asigna el valor a la variable `d` (diámetro).

```python
r = d/2
```
*   **`r = d/2`**:
    *   Calcula el radio de la columna dividiendo el diámetro (`d`) por 2.
    *   El resultado se asigna a la variable `r` (radio).

```python
v = 3.1416 * r**2 * h
```
*   **`v = 3.1416 * r**2 * h`**:
    *   Esta línea calcula el volumen de la columna cilíndrica utilizando la fórmula: `V = π * r² * h`.
    *   Se usa `3.1416` como una aproximación de Pi (π).
    *   `r**2` calcula el radio (`r`) elevado al cuadrado.
    *   El resultado del volumen se asigna a la variable `v`.

```python
sacos = v * 9.5
```
*   **`sacos = v * 9.5`**:
    *   Calcula la cantidad total de sacos de cemento necesarios.
    *   Multiplica el volumen calculado (`v`) por `9.5`, que es la cantidad de sacos requerida por metro cúbico de mezcla.
    *   El resultado se asigna a la variable `sacos`.

```python
print("Volumen ≈", round(v, 3), "m3.")
```
*   **`print("Volumen ≈", round(v, 3), "m3.")`**:
    *   Muestra el texto "Volumen ≈" en la consola.
    *   `round(v, 3)` redondea el valor del volumen (`v`) a 3 decimales para una mejor presentación.
    *   Luego muestra la unidad "m3.".

```python
print("Sacos necesarios ≈", round(sacos, 2), "sacos.")
```
*   **`print("Sacos necesarios ≈", round(sacos, 2), "sacos.")`**:
    *   Muestra el texto "Sacos necesarios ≈" en la consola.
    *   `round(sacos, 2)` redondea la cantidad de sacos (`sacos`) a 2 decimales.
    *   Finalmente, muestra la palabra "sacos.".
